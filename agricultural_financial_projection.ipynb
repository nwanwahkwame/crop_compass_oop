```python
import time as time

from abc import ABC, abstractmethod


class FarmResource: # Class for calculating farm resources like labour, fertilizer and overall farm expenses.

    # Recommended industry standard for total clearing an acre of land. Measured in cedis.
    clearing_constant = 1002

    # Recommended industry standard for total labour on an acre of land. Measured in cedis.
    farm_labour_constant = 1008

    # Decrement multiplier.
    # Cost of clearing likely to decrease due to labour conditions in the Middle and North Hemisphere.
    clearing_decrement_multiplier = 0.8 
    
    # Decrement multiplier.
    # Cost of farm labour likely to decrease due to labour conditions in the Middle and North Hemisphere.
    labour_decrement_multiplier = 0.6

    def __init__(self, fertilizer_cost, is_cleared, acre):
        self.is_cleared = is_cleared
        self.__acre = acre
        self.fertilizer_cost = fertilizer_cost

    @property
    def acre(self):
        return self.__acre

    @acre.setter
    def acre(self, value):
        if value <= 0:
            raise ValueError("Farm size must be greater than 0 acres.")
        self.__acre = value

    def determine_if_cleared(self):
        # Determines cost if land is cleared or not.
        if self.is_cleared == 2:
            return FarmResource.clearing_constant * self.acre
        else:
            return 0

    def labour_cost(self):
        # Determines labour cost.
        labour = FarmResource.labour_constant * self.acre
        return labour

    def total_fertilizer_cost(self):
        # Determines fertilizer cost.
        total_cost = self.fertilizer_cost * self.acre
        return total_cost

    def overall_farm_expenses(self):
        # Determines overall farm expenses.
        clearing = self.determine_if_cleared()
        labouring = self.labour_cost()
        total_of_fertilizer = self.total_fertilizer_cost()
        cost = clearing + labouring + total_of_fertilizer
        return cost


class North (FarmResource): # Child class that inherits from Farm resource class based region chosen by farmer.

    def __init__(self, fertilizer_cost, is_cleared, acre):
        super().__init__(fertilizer_cost, is_cleared, acre)

    def determine_if_cleared(self):
        # Determines cost if land is cleared or not.
        if self.is_cleared == 1:
            return (FarmResource.clearing_constant * self.acre) * \
                FarmResource.clearing_decrement_multiplier
        else:
            return 0

    def labour_cost(self):
        # Determines labour cost.
        labour = (FarmResource.farm_labour_constant * self.acre) * \
            FarmResource.labour_decrement_multiplier
        return labour

    def total_fertilizer_cost(self):
        # Determines fertilizer cost.
        total_cost = (self.fertilizer_cost * self.acre) * \
            FarmResource.labour_decrement_multiplier
        return total_cost

    def overall_farm_expenses(self):
        # Determines overall farm expenses.
        clearing = self.determine_if_cleared()
        labouring = self.labour_cost()
        total_of_fertilizer = self.total_fertilizer_cost()
        cost = (clearing + labouring + total_of_fertilizer) * \
            FarmResource.labour_decrement_multiplier
        return cost


class Mid (FarmResource):# Child class that inherits from Farm resource class based region chosen by farmer

    def __init__(self, fertilizer_cost, is_cleared, acre):
        super().__init__(fertilizer_cost, is_cleared, acre)

    def determine_if_cleared(self):
        # Determines cost if land is cleared or not.
        if self.is_cleared == 1:
            return (FarmResource.clearing_constant * self.acre) * \
                FarmResource.clearing_decrement_multiplier
        else:
            return 0

    def labour_cost(self):
        # Determines labour cost.
        labour = (FarmResource.farm_labour_constant * self.acre) * \
            FarmResource.clearing_decrement_multiplier
        return labour

    def total_fertilizer_cost(self):
        # Determines total cost of fertilizer.
        total_cost = (self.fertilizer_cost * self.acre) * \
            FarmResource.clearing_decrement_multiplier
        return total_cost

    def overall_farm_expenses(self):
        # Determines overall farm expenses.
        clearing = self.determine_if_cleared()
        labouring = self.labour_cost()
        total_of_fertilizer = self.total_fertilizer_cost()
        cost = (clearing + labouring + total_of_fertilizer) * \
            FarmResource.clearing_decrement_multiplier
        return cost


class Information: # Information class that stores researched information of all the various crops.

    # Crop_duration.
    crop_duration = {"Sorghum": 3, "Cotton": 5, "Tomato": 4,
                     "Pepper": 4, "Cocoa": 60, "Groundnut": 4}

    # 2 seed_cost.
    seed_cost = {"Sorghum": 80, "Cotton": 132, "Tomato": 120,
                 "Pepper": 160, "Cocoa": 1350, "Groundnut": 40}

    # 3 seed_spacing = {"crop":"spacing"} cm.
    seed_spacing = {"Sorghum": "60 cm by 75 cm", "Cotton": "90 cm by 30 cm", "Tomato": "60 cm by 75 cm",
                    "Pepper": "60 cm by 75 cm", "Cocoa": "300 cm by 300 cm", "Groundnut": "30 cm by 10 cm"}

    yield_per_acre = {"Sorghum": 1000, "Cotton": 250, "Tomato": 8000,
                      "Pepper": 5000, "Cocoa": 200, "Groundnut": 2000}  # In kg.

    # Number of crops on acre.
    num_of_crops_per_acre = {"Sorghum": 21000, "Cotton": 30000, "Tomato": 3000,
                             "Pepper": 3000, "Cocoa": 444, "Groundnut": 4000}  # Number of crops on acre.

    # Irrigation type.
    irrigation_types = {"drip": ["Tomato", "Pepper", "Cotton"], "rainfall": [
        "Sorghum", "Cocoa", "Groundnut"]}

    # Cost in cedis per acre.
    fertilizer_cost_per_acre = {"Sorghum": 700, "Cotton": 1162,
                                "Tomato": 1600, "Pepper": 1004, "Cocoa": 1004, "Groundnut": 200}  # Cedis.

    crop_fertilizer = {"npk": ["Sorghum", "Groundnut",
                               "Cocoa", "Pepper"], "urea": ["Tomato", "Cotton"]}

    # Selling price of crop in cedis per kg.
    price_per_kg = {"Sorghum": 20, "Cotton": 365, "Tomato": 10,
                    "Pepper": 10, "Cocoa": 40, "Groundnut": 50}

    def __init__(self, crop):
        self.crop = crop

    def all_info(self):

        self.irrigation_type = None
        self.fertilizer = None

        # chosen_crop = self.crop
        duration = self.crop_duration.get(self.crop)
        seed_space = self.seed_spacing.get(self.crop)
        price = self.price_per_kg.get(self.crop)
        crop_num = self.num_of_crops_per_acre.get(self.crop)

        for k, v in self.irrigation_types.items():
            if self.crop in v:
                self.irrigation_type = k

        for k, v in self.crop_fertilizer.items():
            if self.crop in v:
                self.fertilizer = k

        return (duration, crop_num, seed_space, price, self.irrigation_type, self.fertilizer)

    def get_seed_info(self):
        seed_cost_price = self.seed_cost.get(self.crop)
        yield_pa = self.yield_per_acre.get(self.crop)
        price_pkg = self.price_per_kg.get(self.crop)

        # seed cost per acre, yield_per_acre, getting the price of crop in cedis per kg
        return (seed_cost_price, yield_pa, price_pkg)

    def get_fertilizer_cost(self):
        fertilizer_cost = self.fertilizer_cost_per_acre.get(self.crop)
        return fertilizer_cost


class CropFunctionality(ABC): # Abstract class that for calculating planting seed cost and potential yield and revenue

    # Abstraction
    @abstractmethod
    def total_seed_cost():
        pass

    @abstractmethod
    def calculate_yield():
        pass

    @abstractmethod
    def calculate_revenue():
        pass


class Crop(CropFunctionality): # Crop class for calculating planting seed cost and potential yield and revenue

    def __init__(self, seed_cost_per_acre, yield_per_acre, price_per_kg, farm_expenses, farm_acre):
        self.seed_cost_per_acre = seed_cost_per_acre
        self.__farm_acre = farm_acre
        self.yield_per_acre = yield_per_acre
        self.price_per_kg = price_per_kg
        self.farm_expenses = farm_expenses

    def total_seed_cost(self):
        return self.seed_cost_per_acre * self.__farm_acre

    def calculate_yield(self):
        return self.yield_per_acre * self.__farm_acre
        # return total_yield

    def calculate_revenue(self):

        total_of_seed = self.total_seed_cost()
        total_yield = self.calculate_yield()

        all_expenses = total_of_seed + self.farm_expenses

        gross_revenue = total_yield * self.price_per_kg
        net_revenue = gross_revenue - all_expenses

        return (net_revenue, gross_revenue, all_expenses, total_of_seed, total_yield)


class Farmer(): # Farmer ( main class that takes input and makes calls to other classes. )

    # Farmer class to hold validations.
    # App name.
    app_name = "CropCompass"

    # Regions in the country.
    country_hemisphere = ["Northern Ghana", "Mid Ghana", "Southern Ghana"]

    # Crops in the North.
    northern_crops = ["Cotton", "Groundnut", "Sorghum", "Tomato"]

    # Crops in the Middle.
    middle_belt_crops = ["Pepper", "Tomato"]

    # Crops in the South.
    southern_crops = ["Cocoa", "Pepper", "Tomato"]

    # Phone number.
    company_number = "0244454647"

    def __init__(self, land_size: int = 1):
        # Initializing instance attributes.
        while True:

            # Welcome message.
            Farmer.welcome_message()
            # self.welcome_message()

            # Name of prospective farmer.
            self.name = ""
            self.name_validation()

            # Name of region or hemisphere.
            self.hemisphere = ""
            self.hemisphere_validation()

            # Name of prospective crop.
            self.crop = ""
            if self.hemisphere == "Northern Ghana":
                self.northern_crop_validation()
            elif self.hemisphere == "Mid Ghana":
                self.middle_crop_validation()
            else:
                self.south_crop_validation()

            # Land size.
            self.land_size = ""
            self.land_size_validation()

            # State of land, cleared or not.
            self.cleared_land = ""
            self.is_land_clear()

            # Presents overview of user's choices.
            # self.fetch_data()
            self.overview()

            # User input confirmation.
            final_check = print(
                "Are you satisfied with your input, or would you like to start all over (y/n)? ")

            while True:
                final_check = input("""'y' - proceed
'n' - start over
Answer: """)
                if final_check.lower() == "y":
                    break
                elif final_check.lower() == "n":
                    break
                else:
                    print("\nPlease select a valid response.\n")

            if final_check == "y":
                break
            else:
                continue

        # Presents financial implications of user's choices.
        self.financial_forecast()

        # Option for further consultancy.
        self.consultancy()

    @staticmethod
    def welcome_message():
        # Prints out a welcome message.
        print("-"*67)
        print(
            f"Welcome to CropCompass, your financial guide to successful farming.")
        print("-"*67)

    def name_validation(self):
        # Returns the name instance attribute.
        user_name = input("\nA. What is your name? ")
        self.name = user_name.title()
        return self.name

    def hemisphere_validation(self):
        # This method validates and stores the hemisphere attribute.
        print("\nHere are the hemispheres within Ghana: ", "\n")
        for i, area in enumerate(Farmer.country_hemisphere):
            print(f"{i + 1}: {area}")

        threshold = Farmer.country_hemisphere

        while True:

            try:
                user_hemisphere = input(
                    "\nB. Where will your prospective farm be located? ")
                if user_hemisphere.isdigit():
                    user_hemisphere = int(user_hemisphere)
                    if 0 < user_hemisphere <= len(Farmer.country_hemisphere):
                        # Stores hemisphere attribute.
                        self.hemisphere = threshold[user_hemisphere - 1]
                        break
                    else:
                        print(f"Select a number from 1 to {len(threshold)}.")
                else:
                    print("Select a number.")

            except Exception:
                print(f"Select a number from 1 to {len(threshold)}.", "\n")

    def northern_crop_validation(self):
        # This method runs if user selects Northern hemisphere.
        chosen_location = self.hemisphere
        print(f"\nYou selected {chosen_location}.")
        print(
            f"\nThese are the crops within the {chosen_location} hemisphere: ", "\n")

        # Enumerate Northern crops.
        for i, crop in enumerate(Farmer.northern_crops):
            print(f"{i + 1}: {crop.capitalize()}")

        threshold = Farmer.northern_crops

        while True:

            try:
                user_crop = input("\nC. What will your prospective crop be? ")
                if user_crop.isdigit():
                    user_crop = int(user_crop)
                    if 0 < user_crop <= len(threshold):
                        # Stores crop attribute if Northern block runs.
                        self.crop = threshold[user_crop - 1]

                        if user_crop == 1:
                            chosen_crop = threshold[0]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 2:
                            chosen_crop = threshold[1]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 3:
                            chosen_crop = threshold[2]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 4:
                            chosen_crop = threshold[3]
                            print(f"\nYou selected {chosen_crop}.")

                        break
                    else:
                        print(
                            f"Select a number from 1 to {len(threshold)}.")
                else:
                    print("Select a number.")

            except Exception:
                print(f"Select a number from 1 to {len(threshold)}.", "\n")

    def middle_crop_validation(self):
        # This method runs if user selects Mid hemisphere.
        chosen_location = self.hemisphere
        print(f"\nYou selected {chosen_location}.")
        print(
            f"\nThese are the crops within the {chosen_location} hemisphere: ", "\n")

        # Enumerate Middle Belt crops.
        for i, crop in enumerate(Farmer.middle_belt_crops):
            print(f"{i + 1}: {crop.capitalize()}")

        threshold = Farmer.middle_belt_crops

        while True:

            try:
                user_crop = input("\nC. What will your prospective crop be? ")
                if user_crop.isdigit():
                    user_crop = int(user_crop)
                    if 0 < user_crop <= len(threshold):
                        # Stores crop attribute if Middle Belt block runs.
                        self.crop = threshold[user_crop - 1]

                        if user_crop == 1:
                            chosen_crop = threshold[0]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 2:
                            chosen_crop = threshold[1]
                            print(f"\nYou selected {chosen_crop}.")

                        break
                    else:
                        print(
                            f"Select a number from 1 to {len(threshold)}.")
                else:
                    print("Select a number.")

            except Exception:
                print(
                    f"Select a number from 1 to {len(threshold)}.", "\n")
                continue

    def south_crop_validation(self):
        # This method runs if user selects Southern hemisphere.
        chosen_location = self.hemisphere
        print(f"\nYou selected {chosen_location}.")
        print(
            f"\nThese are the crops within the {chosen_location} hemisphere: ", "\n")

        # Enumerate Southern crops.
        for i, crop in enumerate(Farmer.southern_crops):
            print(f"{i + 1}: {crop}")

        threshold = Farmer.southern_crops

        while True:

            try:
                user_crop = input("\nC. What will your prospective crop be? ")
                if user_crop.isdigit():
                    user_crop = int(user_crop)
                    if 0 < user_crop <= len(threshold):
                        # Stores crop attribute if Southern block runs.
                        self.crop = threshold[user_crop - 1]

                        if user_crop == 1:
                            chosen_crop = Farmer.southern_crops[0]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 2:
                            chosen_crop = Farmer.southern_crops[1]
                            print(f"\nYou selected {chosen_crop}.")

                        elif user_crop == 3:
                            chosen_crop = Farmer.southern_crops[2]
                            print(f"\nYou selected {chosen_crop}.")

                        break
                    else:
                        print(
                            f"Select a number from 1 to {len(threshold)}.")
                else:
                    print("Select a number.")

            except Exception:
                print(f"Select a number from 1 to {len(threshold)}", "\n")

    def land_size_validation(self):
        # Stores the land size attribute.

        while True:

            try:
                user_land_size = int(
                    input("\nD. How many acres do you plan to cultivate? "))
                if user_land_size >= 1:
                    self.land_size = user_land_size
                    print(f"\nResponse recorded: {self.land_size} acre(s)")
                    break
                else:
                    print("Select a number equal to, or greater than, 1.")

            except ValueError:
                print("Select a number equal to, or greater than, 1")

    def is_land_clear(self):
        # Stores the cleared land attribute.

        while True:

            try:
                print("\nTell us about the state of your land. ", "\n")
                options = ["Yes", "No"]
                for i, option in enumerate(options):
                    print(f"{i + 1}: {option}")

                response = input(
                    "\nE. Is the land cleared and ready for farming? ")

                if response.isdigit():
                    response = int(response)
                    if 0 < response <= len(options):
                        self.cleared_land = options[response - 1]
                        print(
                            f"\nResponse recorded: {self.cleared_land}")
                        break
                    else:
                        print(f"Select a number from 1 to {len(options)}.")
                else:
                    print("Select a number.")

            except Exception:
                print(f"\nSelect a number.", "\n")

    def overview(self):
        # Prints a brief overview of the user's choices.
        print(f"\nThese are the details you have supplied {Farmer.app_name}:")
        print("-"*52)
        print(f"""Name of farmer: {self.name}
Hemisphere: {self.hemisphere}
Crop: {self.crop}
Land size: {self.land_size} acre(s)
Land cleared: {self.cleared_land}""")
        print("-"*52)

    def fetch_data(self):
        # Allows a brief, three-second reprieve.
        count = 0
        threshold = 2
        print("\nFetching your data. Please wait.")
        while count < threshold:
            for second in range(threshold):
                print(f".", end="")
                time.sleep(1)
            count += 1

    def financial_forecast(self):
        while True:

            confirmation = input(
                "\nWould you like to receive a financial forecast (y/n)? ")
            if confirmation.lower() == "y":
                self.fetch_data()
                print("")
                print("-"*31)
                print("Here's your financial forecast.")
                print("-"*31)
                self.retrieving_information()
                break
            elif confirmation.lower() == "n":
                print("")
                break
            else:
                print("\nPlease select a valid response.")

    # Other class calls, from Information, Crop and North and Mid Ghana classes.

    def retrieving_information(self):

        info = Information(self.crop)
        seed_cpa, yield_pa, price_pkg = info.get_seed_info()

        farm = None

        if self.hemisphere == "Northern Ghana":
            farm = North(info.get_fertilizer_cost(),
                         self.cleared_land, self.land_size)

        elif self.hemisphere == "Mid Ghana":
            farm = Mid(info.get_fertilizer_cost(),
                       self.cleared_land, self.land_size)

        elif self.hemisphere == "Southern Ghana":
            farm = FarmResource(info.get_fertilizer_cost(),
                                self.cleared_land, self.land_size)

        farm.determine_if_cleared()
        farm.labour_cost()
        final_fertilizer = farm.total_fertilizer_cost()
        all_expenses = farm.overall_farm_expenses()

        chosen_crop = Crop(seed_cpa, yield_pa, price_pkg,
                           all_expenses, self.land_size)

        _total_seed_cost = chosen_crop.total_seed_cost()
        _total_yield = chosen_crop.calculate_yield()
        after_revenue = chosen_crop.calculate_revenue()
        other = info.all_info()

        net_revenue, gross_revenue, all_expenses, _total_seed_cost, _total_yield = after_revenue
        duration, crop_num, seed_space, price, irrigation_type, fertilizer = other

        print(f"You chose the {self.crop} crop.\n")
        print(f"The duration of {self.crop} is {duration} months.\n")
        print(
            f"The total cost of seeds for the {self.land_size} acre(s) of land is GH₵ {_total_seed_cost:.2f}.\n")
        print(f"The seed spacing for the crop is {seed_space}.\n")
        print(
            f"The total yield for your {self.land_size} acre(s) of land is {_total_yield} kg.\n")
        print(
            f"The total number of crops will be {self.land_size*crop_num}.\n")
        print(
            f"The irrigation type for your crop will be {irrigation_type}.\n")
        print(f"The fertilizer needed for your crop is {fertilizer}.\n")
        print(
            f"The total cost of fertilizer needed for you farm will be GH₵ {final_fertilizer:.2f}.\n")
        print(
            f"The price per kg for the crop after harvest will be GH₵ {price:.2f}.\n")
        print(f"All expenses will cost you GH₵ {all_expenses:.2f}.\n")
        print(f"Your gross revenue is GH₵ {gross_revenue:.2f}.\n")
        print(f"The profit that you will make is GH₵ {net_revenue:.2f}.\n")

    @classmethod
    def consultancy(cls):
        # Inquiries for further consultancy.
        yes_statement = f"\nFor further financial projections, please call us on: \n{cls.company_number}"
        no_statement = f"\nShould our services be required, do not hesitate to reach out."

        while True:

            consultancy_response = input(
                "\nWould you want more expert opinion (y/n)? ")
            if consultancy_response.lower() == "y":
                print(yes_statement)
                break
            elif consultancy_response.lower() == "n":
                print(no_statement)
                break
            else:
                print("\nPlease select a valid response.")

    def __str__(self):
        # A string representation of the class. It appears last.
        return f"\nThank you, {self.name}, for patronizing {Farmer.app_name}."


farmer = Farmer()
print(farmer)
```
